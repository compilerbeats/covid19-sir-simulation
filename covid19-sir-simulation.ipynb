```python
import random
import math

n = 1000000
```

```python
# init matrix with tupels
# (id, (x_coordinate, y_coordinate), current_state, next_state)

# matrix = np.reshape(np.arange(0,n),(1000,1000))

sqrt_n = int(math.sqrt(n))

matrix = list()
for i in range(0, sqrt_n):
    line_of_matrix = list()
    for j in range(0, sqrt_n):
        x = round(random.uniform(0, sqrt_n), 2)
        y = round(random.uniform(0, sqrt_n), 2)
        line_of_matrix.append(((i * sqrt_n) + j, (x, y), 'S', ''))
    matrix.append(line_of_matrix)

print(matrix[0])
```
