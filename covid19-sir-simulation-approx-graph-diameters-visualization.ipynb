```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("covid19-sir-simulation-approx-graph-diameters.csv", sep=';')

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['alpha'].to_numpy(), df['approx_diameter_max'].to_numpy(), marker='o')

plt.xticks(np.arange(1.1, 3.3, 0.1))
plt.yticks(np.arange(0, 500, 50))

plt.xlabel('Alpha')
plt.ylabel('Approximate diameter')
plt.title('Approximate diameter under alpha')
plt.grid(True)
plt.show()
```
